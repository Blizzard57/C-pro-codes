## Problem 48
> $\sum n^n $ last 10 digits

```python
def prodTen(n):
    val = 1
    for i in range(n):
        val = val * n
        val = val - (val%10**10)
    return val
```

```python
prodTen(99)
```

## Problem 46
> Goldbach's wrong congecture

```python
def isPrime(x):
    for i in range(2,int(x**0.5)+1):
        if x%i == 0:
            return False
    return True

def genPrime(x):
    ls = []
    i = 2
    while True:
        if len(ls) > 0 and ls[-1] >= x:
            return ls
        else:
            if isPrime(i):
                ls.append(i)
            i=i+1
            
def isSquare(n):
    val = n**0.5
    return val - int(val) == 0
```

```python
def goldConj(x):
    ls = genPrime(x)
    if ls[-1] < 0:
        ls = ls[:-1]
    for i in ls:
        if isSquare(i):
            return True
    return False
```

```python
i=2
while True:
    val = goldConj(i)
    #print(i,val)
    if val:
        print(end="")
        #break
    else:
        print("Finally  : ",i,val)
        break
    i+=1
```

```python
def rev(x):
    val = str(x)
    rev = ''
    for i in range(1,len(val)+1):
        rev += val[-i]
        #print(rev)
    fin = int(rev)
    #print(fin)
    return fin

def isPalindrome(x):
    return x == rev(x)

def recPal(x):
    for i in range(50):
        val = x + rev(x)
        if isPalindrome(val):
            return True
        else:
            x = val
    return False
```

```python
val = 0
for i in range(10**4 + 1):
    if recPal(i):
        continue
    else:
        val += 1
```

```python
## Problem 74
def breakNo(x):
    ls = []
    for i in str(x):
        ls.append(int(i))
    return ls

def appFac(ls):
    val = 0
    for i in ls:
        val += math.factorial(i)
    return val

def totFac(x):
    flag = 1
    ls = [x]
    ls.append(appFac(breakNo(ls[-1])))
    while ls[-1] not in ls[:-1]:
        ls.append(appFac(breakNo(ls[-1])))
        flag = flag + 1
        if flag == 60:
            break
    return flag
```

```python
appFac(breakNo(169))
```

```python
ans = 0
for i in range(10**6):
    if totFac(i) == 60:
        ans = ans + 1
```

```python
## Problem 407
def mod(i,x):
    return i%x

def idem(x):
    maxVal = 0
    for i in range(x):
        sq = i**2%x
        #print(i,sq)
        if sq == i:
            if sq > maxVal :
                maxVal = sq
    return maxVal
```

```python
val = 0
for i in range(10**7 + 1):
    val = val + idem(i)
```

```python
## Problem 92
def breakNo(x):
    ls = []
    for i in str(x):
        ls.append(int(i))
    return ls

def sumSq(ls):
    val = 0
    for i in ls:
        val = val + i**2
    return val

def endSeq(x):
    ls = [x]
    while 2 != 3:
        ls.append(sumSq(breakNo(ls[-1])))
        if ls[-1] == 1:
            #print(ls)
            return 0
        elif ls[-1] == 89:
            #print(ls)
            return 1
    return 0
```

```python
val = 0
for i in range(10**7 + 1):
    val = val + endSeq(i)
```
